<h4> imputacion de datos numericos</h4>

```python
def limpia (variable, dataBase):
    dataBase[variable] = pd.to_numeric(dataBase[variable], errors= "coerce")
```

```python
def promedioInt (variable):
    average = int(dataBase[variable].astype("float").mean(axis=0))
    return average
```

```python
def promedioFloat (variable):
    average = round(dataBase[variable].astype("float").mean(axis=0),2)
    return average
```

```python
def modaDate (variable, dataBase):
    moda = dataBase[variable].mode()[0]
    return moda
```

```python
def imputacionNumerica (variable, valor, dataBase):
    if variable in dataBase.columns:
        dataBase[variable].replace(np.nan, valor, inplace=True)
        print("imputacion realizada")
    else:
        print(f"la columna {variable} no existe en el  marco de datos actual")
```
