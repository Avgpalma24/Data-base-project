<h4> imputacion de datos numericos</h4>

```python
def limpia (variable, dataBase):
    dataBase[variable] = pd.to_numeric(dataBase[variable], errors= "coerce")
```

```python
def promedioInt (variable):
    average = int(dataBase[variable].astype("float").mean(axis=0))
    return average
```

```python
def promedioFloat (variable):
    average = round(dataBase[variable].astype("float").mean(axis=0),2)
    return average
```

```python
def modaDate (variable, dataBase):
    moda = dataBase[variable].mode()[0]
    return moda
```

```python
def imputacionNumerica (variable, valor, dataBase):
    if variable in dataBase.columns:
        dataBase[variable].replace(np.nan, valor, inplace=True)
        print("imputacion realizada")
    else:
        print(f"la columna {variable} no existe en el  marco de datos actual")
```

```python
limpia ("list_year", dataBase)
limpia ("assessed_value", dataBase)
limpia ("sale_amount", dataBase)
limpia ("sales_ratio", dataBase)
limpia ("serial_number", dataBase)
```

```python
imputacionNumerica ("serial_number", promedioInt ("serial_number"), dataBase)
imputacionNumerica ("list_year", ( promedioInt ("list_year")), dataBase)
imputacionNumerica ("assessed_value", promedioInt ("assessed_value"), dataBase)
imputacionNumerica ("sale_amount", promedioInt ("sale_amount"), dataBase)
imputacionNumerica ("sales_ratio", promedioFloat ("sales_ratio"), dataBase)
```

<h4 Style = "color: #25857d">1) cual es el ingreso anual de las propiedades vendidas:</h4>

```python
mpl.rcParams["axes.spines.top"]=False
mpl.rcParams["axes.spines.bottom"]=True
mpl.rcParams["axes.spines.left"]=True
mpl.rcParams["axes.spines.right"]=False
```

```python
def agrupaYcuenta (dfParam, nomColumna, orden):
    gf = dfParam.groupby(dfParam[nomColumna].dt.year)["sale_amount"].sum()
    gf = gf.sort_values(ascending=orden)
    return gf
```

```python
dfAgrupado = agrupaYcuenta (dataBase, "date_recorded", False)
```

```python
fig, ax = plt.subplots(1,1, figsize = (15, 6.9), dpi=100) # 

etiquetaBarras= dfAgrupado.index.tolist()# nombre de cada barra en x
alturaBarras = dfAgrupado.tolist()
coloresBarras = ["#25857d","#942222"]

gBarra = ax.bar(etiquetaBarras, alturaBarras, color=coloresBarras)

#-----------------------------------------CONFIG. DEL GRAFICO-----------------------------------------------------------------------#
titulo = "Grafico 1: ingreso anual de las propiedades vendidas"
plt.suptitle(titulo,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = 'bold',
            fontfamily = 'serif')

subtitulo = "Ingreso anual agrupado segun el a√±o"
ax.set_title(subtitulo,
            ha = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 12,
            fontweight = 'regular',
            fontfamily = 'serif')

etiquetax = "A√±os"
ax.set_xlabel(etiquetax,
             ha = 'left',
             fontsize = 10,
             fontweight = 'regular',
             fontfamily = 'serif')

etiquetay = "ingreso anual"
ax.set_ylabel(etiquetay,
             ha = 'left',
             fontsize = 10,
             fontweight = 'regular',
             fontfamily = 'serif')

ax.xaxis.set_label_coords(x=0, y = - 0.07, transform = ax.transAxes)
ax.yaxis.set_label_coords(x=-0.12, y = 0.5, transform = ax.transAxes)

#---------------------------------------------------------------------------------------------#
#-------------------------------Crear etiqueta para cada barra--------------------------------#
#---------------------------------------------------------------------------------------------#
for barra in gBarra:
    altura = barra.get_height()
    ax.text(barra.get_x() + barra.get_width() / 2, altura, str(altura), ha='center', va='bottom', fontsize=10, color='black')
#---------------------------------------------------------------------------------------------#
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.subplots_adjust(top= 0.75)
plt.show()
```
